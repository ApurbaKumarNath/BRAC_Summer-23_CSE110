Assignment 4 (Different methods)

```python
# Task 10
x = input()
x1, x2 = x.split(', ')

for p, q in zip(x1, x2):
  print(p + q, end = '')

if len(x1) < len(x2):
  print(x2[len(x1):])

else:
  print(x1[len(x2):])
```

```python
# Another method for task 10
x = input()
x1, x2 = x.split(', ')

if len(x1) <= len(x2):
  for i in range(len(x1)):
    print(x1[i]+x2[i], end = '')
  print(x2[len(x1):])

else:
  for i in range(len(x2)):
    print(x1[i]+x2[i], end = '')
  print(x1[len(x2):])
```

```python
# Task 19 (Not the exact answer)
in0 = input()
for i in range(len(in0)):
  if i % 2 == 0:
    if 96 < ord(in0[i]) < 123:
      print(chr(ord(in0[i]) - 32), end = '')
    else:
      print(in0[i], end = '')
  else:
    if 64 < ord(in0[i]) < 91:
      print(chr(ord(in0[i]) + 32), end = '')
    else:
      print(in0[i], end = '')
```
