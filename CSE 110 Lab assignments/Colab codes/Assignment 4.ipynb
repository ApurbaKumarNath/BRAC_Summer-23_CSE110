Task 3:
Write a program that takes a string as input and prints “Binary Number” if the string contains only 0s or 1s. Otherwise, print “Not a Binary Number”.

```python
x = input()
binary = True

for i in x:

  if i != '0' and i != '1':
    binary = False
    print('Not a Binary Number')
    break

if binary == True:
  print('Binary Number')
```

Task 10

```python
x = input()
x1 = ''

for i in x:
  if i == ',':
    break
  else:
    x1 += i

x2 = x[len(x1) + 2: len(x)]

if len(x1) >= len(x2):
  for index in range(len(x2)):
    print(x1[index] + x2[index], end = '')
  print(x1[len(x2): len(x1)])

else:
  for index in range(len(x1)):
    print(x1[index] + x2[index], end = '')
  print(x2[len(x1): len(x2)])
```

```python
string = input()
letter = input()
sum = ''

if letter in string:
  for i in string:
    if i == letter:
      continue
    else:
      sum += i
  print(sum)

elif len(string) > 3:
  print(string[1 : len(string)-1])

else:
  print(string)
```
