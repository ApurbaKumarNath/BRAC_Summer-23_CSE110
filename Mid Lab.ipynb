```python
s = input()
for i in s:
  if i == 'a' or i == 'e' or i == 'o' or i == 'u' or i == 'i' or i == 'A' or i == 'E' or i == 'O' or i == 'U' or i == 'I':
    continue
  elif i == '0' or i == '1' or i == '2' or i == '3' or i == '4' or i == '5' or i == '6' or i == '7' or i == '8' or i == '9':
    print(ord(i), end = '')
  else:
    if 65 <= ord(i) <= 89:
      print(chr(ord(chr(ord(i) + 32))+1), end = '')
    else:
      print(chr(ord(i) + 1), end = '')
```
