```python
class MangoTree:
  def __init__(self, variety):
    self.variety = variety
    self.height = 1
    self.number_of_mangoes = 0

  def growthUpdate(self, year):
    self.height += 3*year
    if self.variety == 'Amrapali':
      self.number_of_mangoes += self.height * 15
    elif self.variety == 'Gopalbhog':
      self.number_of_mangoes += self.height * 10
    print(f'Variety: {self.variety}\nHeight: {self.height} meter(s)\nNumber of mangoes on the tree: {self.number_of_mangoes}')

mangoTree1= MangoTree("Gopalbhog")
# Display the details of the mango tree
print("=====================================")
print("Mango Tree Details:")
print(f"Variety: {mangoTree1.variety}")
print(f"Height: {mangoTree1.height} meter(s)")
print(f"Number of mangoes on the tree: {mangoTree1.number_of_mangoes}")
print("=====================================")
mangoTree2= MangoTree("Amrapali")
# Display the details of the mango tree
print("Mango Tree Details:")
print(f"Variety: {mangoTree2.variety}")
print(f"Height: {mangoTree2.height} meter(s)")
print(f"Number of mangoes on the tree: {mangoTree2.number_of_mangoes}")
print("=====================================")

print("Updated details after 5 years:")
mangoTree1= MangoTree("Gopalbhog")
print("=====================================")
mangoTree1.growthUpdate(5)
print("=====================================")
mangoTree2= MangoTree("Amrapali")
mangoTree2.growthUpdate(5)
print("=====================================")
```

```python
class Customer:
  def __init__(self):
    print('Welcome to ABC Memorial Park')
    self.count = 0 # ticket counter
    self.price = 0
  def buyTicket(self, name, age):
    if self.count + 1 > 3:
      print("You can't buy more than 3 tickets ")
    else:
      self.count += 1
      print(f'Successfully purchased a ticket for {name}!')
      if age > 10:
        self.price += 100
      else:
        self.price += 50

  def showDetails(self):
    print('Amount of tickets:', self.count)
    print('Total price:', self.price, 'Taka')

print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
class GreenPhone:
  def __init__(self, model, android, camera):
    self.model = model
    self.series = model[0]
    self.android = android
    self.camera = camera

    if self.series == 'A':
      self.max_android = self.android + 2
    elif self.series == 'M':
      self.max_android = self.android + 3
    elif self.series == 'U':
      self.max_android = self.android + 4

  def showSpecification(self):
    print('Phone Company: GreenPhone')
    print('Model Name:', self.model)
    print('Android Version:', self.android)
    print('Number of Cameras:', self.camera)

  def updatePhone(self):
    if self.android + 1 > self.max_android:
      print(f'Your phone Greenphone {self.model} is already up to date.')
    else:
      self.android += 1
      print(f'Your phone Greenphone {self.model} is upgraded to Android Version: {self.android}.')


print('1=======================')
p1 = GreenPhone('A1', 12, 3)
p2 = GreenPhone('M11', 12, 4)
p3 = GreenPhone('U20', 12, 5)
p1.showSpecification()
print('2=======================')
p2.showSpecification()
print('3=======================')
p1.updatePhone()
print('4=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('5=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('6=======================')
p2.updatePhone()
p3.updatePhone()
print('7=======================')
p1.showSpecification()
p3.showSpecification()
```

```python
class abcTech:
  def __init__(self, name, desig, departm):
    print(f'Welcome to abcTech, {name}!')
    self.name = name
    self.designation = desig
    self.department = departm
    self.skill = []
    self.frame = []

  def addProgrammingSkills(self, lis):
    self.skill += lis

  def addFrameworks(self, li):
    self.frame += li

  def printInfo(self):
    print('Name: ', self.name)
    print('Designation: ', self.designation)
    print('Department:', self.department)
    print('Programming Skills: ', str(self.skill)[1:-1])
    print('Frameworks: ', str(self.frame)[1:-1])

  def calculateSalary(self, base, hour):
    if hour <= 144:
      return base
    else:
      return base + (hour - 144)*800

print("-------------------------")
b1 =abcTech("Tamim Hasan", "Software Engineer", "Android Development")
print("-------------------------")
b1.addProgrammingSkills(["Java", "Python"])
b1.addProgrammingSkills(["Dart", "C++"])
b1.addFrameworks(["Express.js", "React"])
b1.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b1.calculateSalary(45000, 156)}")
print("-------------------------")
print("-------------------------")
b2 =abcTech("Jahin Khandoker", "Senior Developer", "App Development")
print("-------------------------")
b2.addProgrammingSkills(["Java", "Dart", "Swift"])
b2.addFrameworks(["Flutter", "React Native"])
b2.addFrameworks(["Xamarin"])
b2.printInfo()
print("-------------------------")
print(f"Your salary for this month is Tk. {b2.calculateSalary(103000, 123)}")
print("-------------------------")
```

```python
class StudentDatabase:
  def __init__(self, name, id):
    self.name = name
    self.id = id
    self.grades = {}

  def calculateGPA(self, lis, year):
    tup = [] # tup = course names
    grades = 0 # overall grades
    for course in lis:
      tup.append(course[0:6])
      grades += float(course[8:])*3
    tup = tuple(tup)
    gpa = round(grades/(len(tup)*3), 2) # avg grades

    self.grades[year] = {tup: gpa}

  def printDetails(self):
    print('Name:', self.name)
    print('ID:', self.id)
    for year, course_gpa in self.grades.items():
      print(f'Courses taken in {year}:')
      for tup in course_gpa:
        for course in tup:
          print(course) # prints courses
        print(course_gpa[tup]) # prints gpa


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()
```

```python
class Store:
  def __init__(self, balance = 0, items = {}):
    print('New branch created!')
    self.balance = balance
    self.items = items
    self.total_items = len(self.items)


  def viewAllItems(self):
    if self.items == {}:
      print('There are no items in your inventory')
    else:
      s = ''
      for items in self.items:
        s += f'{items}, '
      print('All Items:', s[0: -2])


  def viewAllItemDetails(self):
    print(self.items)

  def add_item(self, lis):
    #lis = ['item', 'stock', 'buying_price', 'selling_price']
    if self.balance >= lis[1]*lis[2]:
      self.items[lis[0]] = {'stock': lis[1], 'buying_price': lis[2], 'selling_price': lis[3]}
      print('Item added:', lis[0])
      self.balance -= lis[1]*lis[2]
      self.total_items += 1

  def sell_item(self, item, quantity):
    if quantity > self.items[item]['stock']:
      print(f'Sorry! {item} is not available at your desired quantity. Currently we have:', self.items[item]['stock'])
    else:
      self.items[item]['stock'] -= quantity
      self.balance += quantity * self.items[item]['selling_price']


  def restock_item(self, item, stock):
    if self.balance >= stock * self.items[item]['buying_price']:
      self.items[item]['stock'] += stock
      self.balance -= stock * self.items[item]['buying_price']
      print(f"Restocked item: {item}, Current Stock: {self.items[item]['stock']}")







print("==========================")
branch1 = Store(5000)
print(f"Current Balance: {branch1.balance}")
print(f"Total items: {branch1.total_items}")
branch1.viewAllItems()
branch1.viewAllItemDetails()
print("==========================")
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["ChaCha Noodles", 10, 5, 8])
print(f"Current Balance: {branch1.balance}")
branch1.add_item(["Sparrow Shampoo", 5, 10, 20])
print(f"Current Balance: {branch1.balance}")
print("==========================")
branch1.viewAllItems()
print()
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.sell_item("ChaCha Noodles", 15)
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
branch1.sell_item("ChaCha Noodles", 10)
print()
print(f"Current Balance: {branch1.balance}\n")
branch1.viewAllItemDetails()
print()
print("==========================")
print(f"Current Balance: {branch1.balance}\n")
branch1.restock_item("ChaCha Noodles", 5)
print()
branch1.viewAllItemDetails()
print()
print(f"Current Balance: {branch1.balance}\n")
print("==========================")
```
