```python
def BMI(height, weight):
  bmi = weight / (height/100)**2
  bmi = round(bmi, 1)
  if bmi < 18.5:
    result = 'Underweight'
  elif 18.5 <= bmi <= 24.9:
    result = 'Normal'
  elif 25 <= bmi <= 30:
    result = 'Overweight'
  elif bmi > 30:
    result = 'Obese'
  return f'Score is {bmi}. You are {result}'
print(BMI(175, 96))
print(BMI(152, 48))
```

```python
def replace_domain(email, new, old = 'kaaj.com'):
  email = email.split('@')
  if email[1] == new:
    return f'Unchanged: {email[0]}@{email[1]}'
  else:
    email[1] = new
    return f'Changed: {email[0]}@{email[1]}'
print(replace_domain('alice@kaaj.com', 'sheba.xyz', 'kaaj.com'))
print(replace_domain('bob@sheba.xyz', 'sheba.xyz'))
```

```python
def palindrome_string(s):
  if ' ' in s:
    s = s.split(' ')
    s = s[0] + s[1]
  if s == s[-1::-1]:
    return 'Palindrome'
  else:
    return 'Not a palindrome'
print(palindrome_string('madam'))
print(palindrome_string('hello'))
print(palindrome_string('nurses run'))
```

```python
def days(day): # days to year month and days
  year = day // 365
  rem_day = day % 365
  month = rem_day // 30
  rem_day = rem_day % 30
  return f'{year} years, {month} months and {rem_day} days'
print(days(4320))
print(days(4000))
```

```python
def correct_capitalization(s):
  li = list(s)
  if 96 < ord(li[0]) < 123:
    li[0] = chr(ord(li[0]) -32)
  for idx in range(0, len(li)):
    if li[idx] in '.!?' and li[idx] == li[-1]:
      break
    if li[idx] == '.' or li[idx] == '?' or li[idx] == '!':
      if 96 < ord(li[idx + 2]) < 123:
        li[idx + 2] = chr(ord(li[idx + 2]) -32)

    if li[idx] == 'i' and li[idx - 1] == ' ' and li[idx + 1] ==' ':
      li[idx] = chr(ord(li[idx]) -32)

    if li[idx] == 'i' and li[idx - 1] == ' ' and li[idx + 1] in '.!?':
      li[idx] = chr(ord(li[idx]) -32)
    string = ''
    for ch in li:
      string += ch
  return string

print(correct_capitalization('my favorite animal is a dog. a dog i. Do i?'))
```

```python
def hospital_fee(**pa): # pa = dictionary of person & amount. key = person, value = amount
  temp = list(pa.values())
  max = temp[0]
  for val in temp:
    if max < val:
      max = val

  payer = ''
  for key, val in pa.items():
    if max == val:
      if payer in pa.keys():
        payer = f'{payer}, {key}'
      else:
        payer = key
  print(f'Highest fee was {max} tk which was paid by {payer}.')
  return payer, max

max_amount, max_payer = hospital_fee(Neymar = 1000, Dembele = 600, Reus = 500, Bale = 1000)
max_amount, max_payer = hospital_fee(Mashrafe = 400, Bumrah = 900, Steyn = 1200, Cummins = 900, Wood = 400, Marsh = 700)
```

```python
def is_james_bond(li): # li = list of integers
  s = ''
  for i in li[-1::-1]:  # looping right to left helps cz you need not consider how many 0's are at left
    if s == '700':
      print(True)
      return True
      break
    elif i in [0, 0, 7]:
      s += str(i)
  if s != '700':
    print(False)
    return False
is_james_bond( [1, 2, 4, 0, 0, 7, 5] )
is_james_bond( [1, 7, 2, 0, 4, 5, 0] )
is_james_bond( [1, 0, 2, 0, 4, 7, 5] )
```

```python
def sec_returner(name, sections):
  sum = 0
  for ch in name:
    sum += ord(ch)
  idx = sum % len(sections)
  sec = sections[idx]
  return sec

def assign_students_to_sections(sections, *students):
  dic = {}
  for sec in sections:
    for std in students:
      if sec == sec_returner(std, sections):
        if sec in dic.keys():
          dic[sec] += [std]
        else:
          dic[sec] = [std]
    if sec not in dic.keys():
      dic[sec] = []
  print(dic)
  return dic

assign_students_to_sections('ABCDE', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace')
```

```python
first_name, middle_name, last_name, student_id= input("First Name: "), input("Middle Name: "), input("Last Name: "), int(input("Student ID: "))
def username_generator(first_name, last_name, student_id, middle_name = ''):
  f_upper = ''
  for ch in first_name:
    if 96 < ord(ch) < 123:
      f_upper += chr(ord(ch) - 32)
    else:
      f_upper += ch
  student_id = str(student_id)
  username = f_upper[0:3] + middle_name + last_name[len(last_name) - 3:] + '_' + student_id[len(student_id) - 4:]
  return username

print(username_generator(first_name, last_name, student_id, middle_name))
```

```python
def key_generator(*name):
  list_key = []
  for person in name:
    first = person[0]
    if 64 < ord(first) < 91:
      first = chr(ord(first) + 32)
    last = person[-1]
    if 96 < ord(last) < 123:
      last = chr(ord(last) - 32)
    ascii = ''
    for ch in person[-2: -len(person): -1]:
      ascii += str(ord(ch))
    key = first + ascii + last
    list_key.append(key)
  print(f'Encrypted Keys: {list_key}')
  return f'Encrypted Keys: {list_key}'

key_list = key_generator("Alex", "Bob", "Trudy")
```

```python
import random
def playRockPaperScissor(rounds):
  userscore = 0
  compscore = 0
  for round in range(rounds):
    user = input()
    computer = random.choice(['rock', 'paper', 'scissor'])
    print('Computer:', computer)

    if user == 'rock' and computer == 'scissor':
      userscore += 1
    elif computer == 'rock' and user == 'scissor':
      compscore += 1
    elif user == 'paper' and computer == 'rock':
      userscore += 1
    elif computer == 'paper' and user == 'rock':
      compscore += 1
    elif user == 'scissor' and computer == 'paper':
      userscore += 1
    elif computer == 'scissor' and user == 'paper':
      compscore += 1

  if userscore > compscore:
    winner = 'You have won the game!'
  elif compscore > userscore:
    winner = 'Computer has won the game!'
  else:
    winner = "It's a tie."
  return f"Your Score: {userscore}\nComputer's Score: {compscore}\n{winner}"

print(playRockPaperScissor(int(input())))
```
