Task 28

```python
start = int(input())
end = int(input())
factors = 1
sum = 0
prime = 0
perfect = 0
primenum = ''
perfectnum = ''

print(f'Between {start} and {end},')


for i in range(start, end + 1):

  while i >= factors:
    if i % factors == 0:
      sum += factors
    factors += 1


  if sum == i + 1:
    prime += 1
    if prime == 1:
      primenum = primenum + str(i)
    else:
      primenum = primenum + ', ' + str(i)


  if sum - factors + 1 == i:
    perfect += 1
    if perfect == 1:
      perfectnum = perfectnum + str(i)
    else:
      perfectnum = perfectnum + ', ' + str(i)

  factors = 1
  sum = 0

print(f'Found {prime} prime numbers')
print(f'Found {perfect} perfect number')
print('Prime numbers:', primenum)
print('Perfect number:', perfectnum)
```

```python
start = int(input())
end = int(input())
divisor = int(input())
mul = 1
for i in range(start, end + 1):
  while i > 0:
    single = i % 10
    mul *= single
    i = i // 10
  if mul % divisor == 0:
    print(mul, end = ' ')
  mul = 1
```
